# Player Tracking Demo
Detect and visualize player positions from a match video using YOLOv8.

```python
import sys
import os
sys.path.append(os.path.abspath('../src/tracking'))

from detect_players import detect_players
import matplotlib.pyplot as plt
import cv2

video_path = '../data/sample_match.mp4'
positions = detect_players(video_path)

for frame_data in positions[:5]:
    frame_num = frame_data['frame']
    centers = frame_data['centers']
    
    cap = cv2.VideoCapture(video_path)
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_num)
    ret, frame = cap.read()
    cap.release()
    
    plt.figure(figsize=(10,6))
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    x_vals = [c[0] for c in centers]
    y_vals = [c[1] for c in centers]
    plt.scatter(x_vals, y_vals, c='red', s=100, alpha=0.7)
    plt.title(f'Frame {frame_num} - Player Positions')
    plt.axis('off')
    plt.show()
```
